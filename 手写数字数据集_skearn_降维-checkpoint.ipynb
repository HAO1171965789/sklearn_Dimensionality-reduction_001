# 1.降维主要方法介绍
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;机器学习任务主要分为监督学习任务和无监督学习任务。监督学习任务有两大类，分别是分类和回归。无监督学习任务有两大类，聚类和降维。在本案例中我们主要应用降维的方法。降维顾名思义就是降低数据维度，降维的主要目的如下：
<ol>
<li>摒弃冗余信息（共线性），删除噪音 </li> 
<li>降低计算时间、空间复杂度</li>
<li>简单模型具有更好的鲁棒性 </li>
<li>实现数据可视化</li>
<li>可以作为特征提取的方法</li>
</ol>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;降维的主要方法分为基于矩阵分解的、线性的方法和基于保留局部/全局特征的、非线性的方法。

# 2.数据预览及预处理

导入所需要的模块，读入数据集，查看数据。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from time import time
% matplotlib inline
# 导入数据集，为了便于可视化展示，这次我们选取n_class=5，即只选取数字类型为0、1、2、3、4的这五种类别的数据
digits = datasets.load_digits(n_class=5)
X = digits.data
y = digits.target
print(X[:2],'\n')
print(y[:5])
print(len(X))
```

上述对数据的读取结果不能使读者对digits数据有直观的认识，为了加深对digits数据集的理解，下列对全部原始数据进行简单的可视化。

```python
# 定义数组
n_img_per_row = 20
img = np.zeros((10 * n_img_per_row, 10 * n_img_per_row))
print(img.shape)

# 对数组进行循环遍历
for i in range(n_img_per_row):
    ix = 10 * i + 1
    for j in range(n_img_per_row):
        iy = 10 * j + 1
        # 把手写数据X进行维度变换转化为8行8列并依次赋值给img数组的对应位置
        img[ix:ix + 8, iy:iy + 8] = X[i * n_img_per_row + j].reshape((8, 8))
        
plt.imshow(img, cmap=plt.cm.binary)
plt.title('A selection from the 64-dimensional digits dataset')
```

原始数据中有五种类别，为了让读者更清楚`digits.data`，依次对类别为数字0、1、2、3、4的单个`digits.data`数据进行可视化展示。

# 3.使用不同降维方法对数据集进行降维
** 构建绘图函数  **

本案例使用不同降维方法把数据降维至2维或3维，为了便于形象地展示低维数据，我们定义了以下函数，可以将降维后的2维或3维数据进行可视化。

```python
# 定义2维数据可视化函数
def plot_2d(X, title):
    # 归一化 ，在[0,1]区间上展示数据
    x_min, x_max = np.min(X,axis=0), np.max(X,axis=0)
    X = (X - x_min) / (x_max - x_min)
    fig = plt.figure()
    # 依次遍历X的第0、1列作为二维坐标，在二维坐标系对应位置通过plt.text函数添加纯文本注释，注释的值就是当前X数据对应的类别
    for i in range(X.shape[0]):
        plt.text(X[i][0], X[i][1],str(digits.target[i]),
                 color=plt.cm.Set1(y[i] / 5.),
                 fontdict={'size': 9})
    plt.title(title)
```

```python
# 定义3维数据可视化函数
from mpl_toolkits.mplot3d import Axes3D
def plot_3d(X, title):
    # 归一化 ，在[0,1]区间上展示数据
    x_min, x_max = np.min(X,axis=0), np.max(X,axis=0)
    X = (X - x_min) / (x_max - x_min)
    
    fig = plt.figure()
    ax = Axes3D(fig)
    # 依次遍历X的第0、1、2列作为三维坐标，在三维坐标系对应位置通过plt.text函数添加纯文本注释，注释的值就是当前X数据对应的类别
    for i in range(X.shape[0]):
        ax.text(X[i][0], X[i, 1], X[i,2],str(digits.target[i]),
                 color=plt.cm.Set1(y[i] / 5.),
                 fontdict={ 'size': 9})
    plt.title(title)
```

## 3.1PCA降维

主成分分析（PCA）的基本思想是：构造原始特征的一系列线性组合形成低维的特征，以去除数据的相关性，并使降维后的数据最大程度地保持原始高维数据的方差信息。以下通过PCA降维把数据降至3维，并调用函数进行可视化。

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3, random_state=0)
t0 = time()
X_pca = pca.fit_transform(X)
plot_3d(X_pca,"PCA 3D (time %.2fs)" %(time() - t0))
```

由图可以看出分类效果还可以，我们可以分辨出有五种数字类别，但是同类别数据的不够集中，不同类别的数据距离不够远。考虑到能否进一步提升降维的效果，接下来我们引入核方法的主成分分析。核方法与SVM的核方法等含义相同，意为将原始数据投影到更高维空间后在另一个方向投影使特征更明显。
<div id='3.2'></div>
## 3.2KernelPCA降维
数据通过核函数投影，`sklearn.decomposition`中`KernelPCA`提供了五种核函数。我们接下来对着五种核函数的核PCA方法降维进行画图。

```python
from sklearn.decomposition import KernelPCA

ker = ['linear','poly','rbf','sigmoid','cosine']
fig = plt.figure(figsize=(20,20))
axes = [321,322,323,324,325]

for ker_index ,ax in enumerate(axes):
    # 核函数为ker[ker_index]时，经过降维后的可视化3D图形绘制在子图ax处
    ax = fig.add_subplot(ax, projection='3d')
    kpca = KernelPCA(n_components=3,kernel=ker[ker_index],random_state=0)
    
    t0 = time()
    X_kpca = kpca.fit_transform(X)
    # 归一化 ，在[0,1]区间上展示数据
    x_min, x_max = np.min(X_kpca,axis=0), np.max(X_kpca,axis=0)
    X_kpca = (X_kpca - x_min) / (x_max - x_min)
    # 依次遍历X的第0、1、2列作为三维坐标，在三维坐标系对应位置通过plt.text函数添加纯文本注释，注释的值就是当前X数据对应的类别
    for k in range(X.shape[0]):
        ax.text(X_kpca[k][0], X[k, 1], X_kpca[k,2],str(digits.target[k]),
                 color=plt.cm.Set1(y[k] / 5.),
                 fontdict={ 'size': 9})
    plt.title("%sPCA 3D (time %.2fs)" %(ker[i], time() - t0))
```

由上图可看出使用不同核函数的KernelPCA也能使五类样本区分开来，但是较PCA降维来说降维效果并没有明显的改善。

接下来我们使用三种非线性的的降维方法对原始数据进行降维并进行可视化，观察使用基于流形学习的降维方法来进行降维的降维效果，力求找到更好的降维方法。这三种方法分别是多维标度法（MDS），局部线性嵌入（LLE），t分布随机邻域嵌入（t-SNE），让我们开始吧。

<div id='3.3'></div>
## 3.3MDS降维
多维尺度变换能够只利用样本间的距离信息，找到每一个样本的特征表示，且在该特征表示下样本的距离与原始的距离尽量接近。因为多维尺度变换能够在低维特征中保持样本的距离信息。



`sklearn.manifold`中封装了MDS算法，从中导入MDS模块，使用MDS降维方法将数据集降至3维并对3维数据进行可视化。

```python
from sklearn.manifold import MDS
mds = MDS(n_components=3,random_state=0)
t0 = time()
X_mds = mds.fit_transform(X)
plot_3d(X_mds,"MDS 3D (time %.2fs)" %(time() - t0))
```

<div id='3.4'></div>
## 3.4LLE降维

在MDS算法中，我们通过保持高维空间中样本的距离来得到数据的低维表示。然而当数据分布在高维空间中的一个流形上时，仅依靠距离不能反映这个特点。LLE算法假设数据分布在高维空间中的低维流形内，它的基本思想是数据在局部存在线形关系，将数据降维后这种局部的线性关系需要得到保持。

```python
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_neighbors=30,n_components=3,method='standard',random_state=0)
t0 = time()
X_lle = lle.fit_transform(X)
plot_3d(X_lle,"LLE 3D (time %.2fs)" %(time() - t0))
```

## 3.5 TSNE降维

TSNE是由SNE衍生出的一种算法，SNE最早出现在2002年，它改变了MDS和ISOMAP中基于距离不变的思想，将高维映射到低维的同时，尽量保证相互之间的分布概率不变，SNE将高维和低维中的样本分布都看作高斯分布，而Tsne将低维中的坐标当做T分布，这样做的好处是为了让距离大的簇之间距离拉大，从而解决了拥挤问题。从SNE到TSNE之间，还有一个对称SNE，其对SNE有部分改进作用。

`sklearn.manifold`中封装了TSNE算法，从中导入TSNE模块，使用TSNE降维方法将数据集降至3维并对3维数据进行可视化。

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=3,random_state=0)
t0 = time()
X_tsne = tsne.fit_transform(X)
plot_3d(X_tsne,"TSNE 3D (time %.2fs)" %(time() - t0))
```

对比以上所有降维方法，MDS降维后的可视化效果较PCA有所提升，但不是特别明显。TSNE降维后的可视化效果最好，可以看到同类别数据的较为集中，不同类别的数据有明显的距离间隔，其降维效果应该是最好的（具体衡量各种算法的降维效果详见第4章节）。

另外由上图可以看出，数据分布集中在二维平面上，在三维方向的波动不是特别明显。考虑使用TSNE降维方法把数据维数降维至2维，通过2维数据的分布观察降至2维的降维效果。

由二维数据的可视化可以明显看出通过TSNE降维至二维，降维效果依旧突出。

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2,random_state=0)
t0 = time()
X_tsne = tsne.fit_transform(X)
plot_2d(X_tsne,"TNSE 2D (time %.2fs)" %(time() - t0))
```

综上，TSNE将高维映射到低维的同时，尽量保证相互之间的分布概率不变，其对digits数据集降维效果最好。但是TSNE的算法复杂度也是最高的，所耗费的时间也要远远大于其他降维方法。

另外，我们也可以由降维效果看出digits数据集中的每个数据虽然有64维，但是维度之间有很强的相关性，用二维就能很好的表示数据的特性。这也正是降维的意义所在。


# 4.使用逻辑回归判定降维效果
在此章节，我们通过对比原始数据与通过不同降维方法降维后的数据的逻辑回归的效果，可以来量化降维效果。通过模型预测的score值衡量分类效果，进而可以表示不同降维模型的降维效果。score值越高，降维效果越好。以下我们对PCA，MDS，LLE，TSNE这四种降维方法的降维效果做对比分析。

首先导入相关模块。

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

原始数据集(未降维)经LogisticRegression拟合后模型的score值。

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
print('未降维模型的正确率：%s'%(clf.score(X_test,y_test)))
```

通过PCA把原始数据集降至3维，再经LogisticRegression拟合后模型的score值。

```python
X_redu = PCA(n_components=3,random_state=0).fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_redu, y, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
print('经过PCA降维后模型的正确率：%s'%(clf.score(X_test,y_test)))
```

通过MDS把原始数据集降至3维，再经LogisticRegression拟合后模型的score值。

```python
X_redu = MDS(n_components=3,random_state=0).fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_redu, y, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
print('经过MDS降维后模型的正确率：%s'%(clf.score(X_test,y_test)))
```

通过LLE把原始数据集降至3维，再经LogisticRegression拟合后模型的score值。

```python
X_redu = LocallyLinearEmbedding(n_neighbors=30,n_components=3,method='standard',random_state=0).fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_redu, y, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
print('经过LLE降维后模型的正确率：%s'%(clf.score(X_test,y_test)))
```

通过TSNE把原始数据集降至3维，再经LogisticRegression拟合后模型的score值。


```python
X_redu = TSNE(n_components=3,random_state=0).fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_redu, y, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
print('经过TSNE降维后模型的正确率：%s'%(clf.score(X_test,y_test)))
```

由上述score值可以发现TSNE的降维效果明显优于其他方法的降维效果。与原始数据集相比，经过TSNE降维后，模型变得简单，score值也有明显提升。

由上述结果可以看出LLE降维效果相对其他方法不太理想，这与预想不太相符，此时可以考虑通过调节参数来优化降维效果。接下来我们以`n_neighbors`参数为例，找出最优的`n_neighbors`参数。当然若需要使模型达到最佳降维效果，在计算量可以接受的情况下，最好调节尽量多的相关参数寻找参数的最优解。

```python
n_neighbors = [10,20,30,40,50]
for n in n_neighbors:
    X_redu = LocallyLinearEmbedding(n_neighbors=n,n_components=3,method='standard',random_state=0).fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_redu, y, random_state=0)
    clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial',C=10.0).fit(X_train, y_train)
    print('经过LLE-n_neighbors=%s降维后模型的正确率：%s'%(n,clf.score(X_test,y_test)))
```

由上述score值可以明显发现`n_neighbors=10`时LLE的降维效果要明显优于`n_neighbors=30`时的降维效果。

画出LLE降维后的3D可视化图形，从图形也可见`n_neighbors=10`时LLE的降维效果要明显优于`n_neighbors=30`时的降维效果。

```python
lle = LocallyLinearEmbedding(n_neighbors=10,n_components=3,method='standard',random_state=0)
t0 = time()
X_lle = lle.fit_transform(X)
plot_3d(X_lle,"LLE 3D (time %.2fs)" %(time() - t0))
```

# 5.总结
降维的目的是降低维度，便于后续的计算和可视化，其更深层次的意义在于有效信息的提取综合及无用信息的摈弃。为此我们首先要根据数据在高维空间的表现形式来选择合适的降维方法，降维的方法主要为两种：projection 和 Manifold Learning。

在大多数的真实问题，训练样例都不是均匀分散在所有的维度，许多特征都是固定的，同时还有一些特征是强相关的，因此所有的训练样例实际上可以投影（projection）在高维空间中的低维子空间中。然而投影（projection）不总是降维最好的方法在，比如许多情况下，空间可以扭转，如著名的瑞士卷（Swiss roll）数据，瑞士卷（Swiss roll）是二维流形的例子，它可以在高维空间中弯曲。此时就要考虑基于流行数据进行建模的降维算法--流形学习（Manifold Learning），它假设大多数现实世界的高维数据集接近于一个低维流形。

然后要根据降维的效果不断调整参数，这样才能得到数据在低维空间下的最真实的表示。
